```python
# Question 10
dna2dna = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
def reverse_complement(dna):
    """Return a string representing the reverse complement of the single dna strand."""
    other = ''              # start with an empty string
    for base in dna:
        other = other + dna2dna[base]
    return other[ : :-1]

def mutate(dna, motif):
  x= dna.find(motif)
  spliced=[]
  if dna.find(motif)>=0:
    length= len(motif)
    y= dna.find(reverse_complement(motif),x)
    spliced= dna[0:x]+dna[y+length:]
    return(spliced)
  else:
    return(dna)

print(reverse_complement("G"))
mutate('AGTCACATGATCAGT',"C")
```
